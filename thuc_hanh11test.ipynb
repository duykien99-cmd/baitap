```python
# [Date]
# Xây dựng lớp Date biểu diễn ngày tháng năm với
# hai phương thức kiểm tra xem một ngày tháng
# nào đó có hợp lệ hay không và phương thức tính
# ngày kế tiếp của ngày tháng năm nào đó nếu nó là
# hợp lệ
# Nhập vào một ngày tháng năm, (ngày tháng năm
# biểu diễn dưới dạng dd/mm/yyyy). Nếu ngày
# tháng năm không tồn tại, in ra dòng chữ INVALID.
# Nếu là hợp lệ in ra ngày kế tiếp của nó.
# Đầu vào
# Đầu vào từ bàn phim chứa dữ liệu ngày, tháng,
# năm dưới dạng dd/mm/yyyy.
# Đầu ra
# In ra màn hình ngày kế tiếp của ngày vừa nhập
# dưới dạng dd/mm/yyyy. 
# Nếu ngày tháng không tồn tại, in ra INVALID
class Date:
    def __init__(self,day,month,year):
        self.d=day
        self.m=month
        self.y=year
    def nam_nhuan(self):
        return self.y%400==0 or (self.y%4==0 and self.y%100!=0)
    def ngay_cuoi_thang(self):
        c=[1,3,5,7,8,10,12]
        d=[4,6,9,11]
        if self.m==2:
            if self.nam_nhuan():
                return 29
            else:
                return 28
        elif self.m in c:
            return 31
        elif self.m in d:
            return 30
    def is_Valid(self):
        return 1<=self.d<=self.ngay_cuoi_thang() and 1<=self.m<=12
    def ngay_tiep_theo(self):
        if self.d==self.ngay_cuoi_thang():
            self.d=1
            self.m+=1
            if self.m>12:
                self.m=1
                self.y+=1
        else:
            self.d+=1
    def __str__(self):
        if not self.is_Valid():
            return "INVALID"
        else:
            return f"{self.d:02d} / {self.m:02d} / {self.y}"
a,b,c=map(int,input().strip().split("/"))
g=Date(a,b,c)
g.ngay_tiep_theo()
print(g)
```

```python
# Xây dựng lớp điểm với hai thuộc tính x, y cùng
# các phương thức kiểm tra xem có phải gốc tọa dộ
# không, có nằm trên trục tung không, có nằm trên
# trục hoành không, tính khoảng cách từ điểm đó tới
# một điểm khác
# Đầu vào:
# Nhập vào tọa độ một điểm M bất kỳ
# Đầu ra:
# Thông báo M có phải gốc tọa độ không hoặc có
# nằm trên trục tung hoặc trục hoành không, khoảng
# cách từ M tới gốc tọa độ là bao nhiêu (lấy 2 chữ
# số phần thập phân)
import math
class point:
    def __init__(self,a,b):
        self.x=a
        self.y=b
    def nam_tren_truc_hoanh(self):
        return self.x!=0 and self.y==0
    def nam_tren_truc_tung(self):
        return self.x!=0 and self.y==0
    def goc_toa_do(self):
        return self.x==0 and self.y==0
    def khoang_cach_toi_gtd(self):
        return math.sqrt(self.x**2+self.y**2)
    def khong_thuoc_truc_nao(self):
        return self.x!=0 and self.y!=0
    def __str__(self):
        if self.nam_tren_truc_hoanh():
            return (f"nam tren truc hoanh\n"
                    f"{self.khoang_cach_toi_gtd()}")
        elif self.nam_tren_truc_tung():
            return (f"nam tren truc tung\n"
                    f"{self.khoang_cach_toi_gtd()}")
        elif self.khong_thuoc_truc_nao():
            return (f"khong thuoc truc nao\n"
                    f"{self.khoang_cach_toi_gtd()}")
        elif self.goc_toa_do():
            return (f"day la goc toa do\n"
                    f"0")
a,b=map(int,input().split())
c=point(a,b)
print(c)
```

```python
# Lớp BankAccount(owner, balance) với:
# deposit(x), withdraw(x); nếu rút quá số dư thì bỏ
# qua giao dịch.
# __str__() in owner: balance.
# Input
# Dòng 1: owner initial_balance
# Dòng 2: n (số lệnh)
# Tiếp theo n dòng, mỗi dòng: DEPOSIT x hoặc
# WITHDRAW x.
# Output
# Số dư cuối cùng (in số thực, 2 chữ số).
class BankAccount:
    def __init__(self,owner,balance):
        self.x=owner
        self.y=float(balance)
    def deposit(self,so_tien_them):
        y=float(so_tien_them)
        self.y+=y
        return float(self.y)
    def withdraw(self,so_tien_rut):
        x=float(so_tien_rut)
        if x<=self.y:
            self.y-=x
        return float(self.y)
    def __str__(self):
        return f"{self.y:.2f}"
name,balance=input().split()
c=BankAccount(name,balance)
n=int(input())
for i in range(n):
    a,b=input().split()
    if a.lower()=="deposit":
        c.deposit(b)
    elif a.lower()=="withdraw":
        c.withdraw(b)
print(c)
        
    

```

```python
import math
class ComplexNumber:
    def __init__(self,initRealPart,initlmaginaryPart):
        self.a=float(initRealPart)
        self.b=float(initlmaginaryPart)
    def __add__(self,other):
        tong_thuc=self.a+other.a
        tong_ao=self.b+other.b
        return f"{tong_thuc:.2f} + {tong_ao:.2f}i"
    def __mul__(self,other):
        phan_thuc=self.a*other.a-self.b*other.b
        phan_ao=self.a*other.b+self.b*other.a
        return f"{phan_thuc:.2f} + {phan_ao:.2f}i"
    def abs(self,other):
        phan_thuc1=self.a*other.a-self.b*other.b
        phan_ao1=self.a*other.b+self.b*other.a
        return f"{float(math.sqrt(phan_thuc1**2+phan_ao1**2)):.2f}"
    def conjugate(self,other):
        phan_thuc2=self.a*other.a-self.b*other.b
        phan_ao2=self.a*other.b+self.b*other.a
        phep_toan="+" if phan_ao2 < 0 else "-"
        return f"{phan_thuc2:.2f} {phep_toan} {abs(phan_ao2):.2f}i"
    def __str__(self):
        pass
a,b=map(int,input().split())
e=ComplexNumber(a,b)
c,d=map(int,input().split())
g=ComplexNumber(c,d)
print(e.__add__(g))
print(e.__mul__(g))
print(e.abs(g))
print(e.conjugate(g))
    
```

```python
import math
class Triangle:
    def __init__(self,edge1,edge2,edge3):
        self.a=edge1
        self.b=edge2
        self.c=edge3
    def double_getArea(self):
        if self.a >0 and self.b >0 and self.c>0 and (self.a+self.b>self.c) and (self.b+self.c>self.a) and (self.a+self.c>self.b):
            p=(self.a+self.b+self.c)/2
            return f"{math.sqrt(p*(p-self.a)*(p-self.b)*(p-self.c)):.2f}"
        else:
            return "invalid"
a,b,c=map(float,input().split())
d=Triangle(a,b,c)
print(d.double_getArea())
```

```python
class TimeStamp:
    def __init__(self,hour,minute,seconds):
        self.h=hour
        self.m=minute
        self.s=seconds
    def is_valid(self):
        return 0<=self.s<60 and 0<=self.m<60 and 0<=self.h<24
    def sau_do_1s(self):
        if self.s==59:
            self.s=0
            self.m+=1
            if self.m==60:
                self.m=0
                self.h+=1
                if self.h==24:
                    self.h=0
        else:
            self.s+=1
        return f"{self.h:02d}:{self.m:02d}:{self.s:02d}"
    def trc_do_1s(self):
        if self.h==0:
            self.h=23
            self.m-=1
            if self.m<0:
                self.m=59
                self.s-=2
                if self.s<0:
                    self.s=60+int(self.s)
        elif self.h!=0 and self.m==0:
            self.h-=1
            self.m-=1
            if self.m<0:
                self.m=60+int(self.m)
                self.s-=2
                if self.s<0:
                    self.s=60+int(self.s)
        else:
            self.s-=2
        return f"{self.h:02d}:{self.m:02d}:{self.s:02d}"
    def __str__(self):
        if self.is_valid():
            return (f"{self.sau_do_1s()}\n"
            f"{self.trc_do_1s()}")
        else:
            return "INVALID"
a,b,c=map(int,input().strip().split(":"))
d=TimeStamp(a,b,c)
print(d)

        
        
```
