```python
#W4A6
def prime(c):
    if c<2:
        return False
    elif c==2:
        return True
    else:
        for i in range(2,int(c**0.5)+1):
            if c%i==0:
                return False
    return True
a,b=map(int,input().split())
ans=0
for i in range(a,b+1):
    if prime(i):
        ans+=i
print(ans)
```

```python
#W4A8
n=input()
N=int(n)
def palindrome(a):
    if a==a[::-1]:
        return True
    return False
for _ in range(100000):
    if palindrome(n):
        print(n)
    else:
        N+=int(n[::-1])
        if palindrome(str(N)):
            print(N)
            break
        else:
            n=str(N)
```
