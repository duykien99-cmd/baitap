```python
#Viết hàm kiểm tra một số có phải là số nguyên tố hay không?
def songuyento(n):
    if n<2:
        return False
    elif n==2:
        return True
    for i in range(3,n):
        if n%i==0:
            return False
    return True
a=int(input())
if songuyento(a):
    print("day la so nguyen to")
else:
    print("day kh phai so nguyen to")
```

```python
#W5A4: VIết hàm kiểm tra một số có phải số hoàn hảo hay không
# xử lý trên dữ liệu kiểu số nguyên
#số hoàn hảo là số có các ước (trừ chính nó) có tổng bằng nó
def sohoanhao(n):
    a=[]
    if n==1:
        return True
    for i in range(1,(n//2)+1):
      if n%i==0:
            a.append(i)
    if sum(a)==n:
        return True
    return False
e=int(input())
if sohoanhao(e):
    print(e,"la so hoan hao")
else:
    print("day kh phai so hoan hao")
    
            
            
```

```python
n_input = int(input("Nhập số nguyên thứ nhất (n): "))
y_input = int(input("Nhập số nguyên thứ hai (y): "))

# 1. Chuyển đổi sang nhị phân
def convert_to_binary_list(num):
    if num == 0:
        return [0]
    binary_list = []
    temp_num = num
    while temp_num > 0:
        remainder = temp_num % 2
        binary_list.append(remainder)
        temp_num //= 2
    binary_list.reverse()
    return binary_list

a = convert_to_binary_list(n_input)
b = convert_to_binary_list(y_input)

# 2. Đồng nhất độ dài (Padding)
max_len = max(len(a), len(b))
# Thêm số 0 vào đầu (bên trái) chuỗi nhị phân ngắn hơn
a = [0] * (max_len - len(a)) + a
b = [0] * (max_len - len(b)) + b

# 3. Tính khoảng cách Hamming
hamming_distance = 0
# Vòng lặp bây giờ chạy qua độ dài đồng nhất (max_len)
for i in range(max_len):
    if a[i] != b[i]:
        hamming_distance += 1

print(f"Chuỗi nhị phân của {n_input}: {''.join(map(str, a))}")
print(f"Chuỗi nhị phân của {y_input}: {''.join(map(str, b))}")
print(f"Khoảng cách Hamming: {hamming_distance}")
```
