```python
#Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng
#giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.
#Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]
n=map(int,input().split())
a=set(n)
b=[]
for i in a:
    if i not in b:
        b.append(i)
print(b)
```

```python
#Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu
#tới vị trí i), giữ nguyên thứ tự
a=map(int,input().split())
b=list(a)
tong=0
c=[]
for i in b:
    tong+=i
    c.append(tong)
print(c)
    
```

```python
#Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple 
#sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp 
#xếp.)
n=map(int,input().split())
k=int(input())
a=list(n)
for _ in range(k):
    a.append(a[0])
    a.remove(a[0])
c=tuple(a)
print(c)
```

```python
#Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo 
#dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key 
#lặp lại, thêm vào list của key đó 
##Tách chuỗi i bằng( dấu hai chấm (:) là dấu giúp python nhận biết là tách cặp key,value ) hai phần tử, sau đó gán phần tử đầu tiên cho key và phần tử thứ hai cho value 
#là lệnh i.split(':')
n=input()
m=n.split()
dic={}
for i in m:
    key,value=i.split(':')
    if key not in dic:
        dic[key]=[value]
    else:
        dic[key].append(value)
print(dic)



```

```python
#Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 
#'zeros': ...} (đếm số >0, <0, =0)
n=list(map(int,input().split()))
a={'positives':[],'negatives':[],'zeros':[]}
d1=0
d2=0
d3=0
for i in n:
    if i>0:
        a['positives'].append(i)
    elif i==0:
        a['zeros'].append(i)
    else:
        a['negatives'].append(i)
print(len(a['positives']),len(a['zeros']),len(a['negatives']))
```

```python
#Viết chương trình nhập vào một list các chuỗi, sau đó đếm số lần xuất hiện 
#của từng chuỗi và lưu vào dictionary. 
#for i in m là chạy từng list con trong 1 list lớn và các list coc cách nhau bởi 1 dấu cách
n=input()
m=n.split()
dic={}
for i in m:
    if i not in dic:
        dic[i]=1
    else:
        dic[i]+=1
print(dic)



    


        
```

```python
#Nhập vào một tuple chứa các số nguyên. Hãy tạo ra hai tuple mới: một chứa 
#tất cả các số chẵn và một chứa tất cả các số lẻ từ tuple ban đầu, giữ nguyên 
#thứ tự tương đối. In ra hai tuple này trên hai dòng riêng biệt. 
n=map(int,input().split())
a=[]
c=[]
for i in n:
    if i%2==0:
        a.append(i)
    else:
        c.append(i)
print(tuple(a))
print(tuple(c))
```

```python
#Viết chương trình nhập vào một list các chuỗi, sau đó đếm số lần xuất hiện 
#của từng chuỗi và lưu vào dictionary.
n=input()
m=n.split()
a={}
for key in m:
    if key not in a:
        a[key]=1
    else:
        a[key]+=1
print(a)
```

```python
#Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong
#list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ
#nhất trong số đó.
b=input()
a=b.split()
dic_a={}
max=0
for i in a:
    if i not in dic_a:
        dic_a[i]=1
    else:
        dic_a[i]+=1
for i in dic_a:
    if dic_a[i]>max:
        max = dic_a[i]
        ans=i
    elif dic_a[i]==max:
        if i<ans:
            ans=i          
print(ans)
```

```python
#Nhập vào hai dòng, mỗi dòng là một danh sách các số nguyên. Hãy tìm các
#phần tử chung (giao) của hai danh sách này và in ra một list mới chứa các
#phần tử chung đó, không trùng lặp.
m=list(map(int,input().split()))
n=list(map(int,input().split()))
c=[]
for i in m:
    if i in n:
        c.append(i)
print(c)





```

```python
#Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một
#dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ
#được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới
#theo thứ tự từ điển tăng dần của key.
n=input()
m=n.split()
dic={}
for i in m:
    key,value=i.split(":")
    dic[key]=value
n1=input()
m1=n1.split()
dic1={}
for i in m1:
    key1,value1=i.split(":")
    dic1[key1]=value1
for key1 in dic:
    if key1 not in dic:
        dic[key1].append(value1)
    else:
        dic[key1]+=dic1[key1]
print(dic)

```

```python
#[Đảo ngược key và value của dictionary]
#Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo
#một dictionary mới bằng cách đảo ngược key và value của dictionary ban
#đầu. In ra dictionary mới.
n=input()
m=n.split()
dic={}
for i in range(0,len(m),2):
    key,value=m[i],m[i+1]
    dic[value]=key
print(dic)
    
    
```

```python
#Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số
#nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp
#key-value từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary mới.
n=input()
m=n.split()
k=int(input())
dic={}
for i in range(0,len(m),2):
    key,value=m[i],m[i+1]
    if int(value)>k:
        dic[key]=value
print(dic)
```
