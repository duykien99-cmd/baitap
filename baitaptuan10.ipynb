```python
# Viết hàm merge hai danh sách đã sắp xếp mà không dùng bất cứ
# thuật toán sắp xếp nào
def merge(a,b):
    c=[]
    j=0
    i=0
    while i<len(a) and j<len(b):
        if a[i] < b[j]:
            c.append(a[i])
            i+=1
        else:
            c.append(b[j])
            j+=1
    c.extend(a[i:])
    c.extend(b[j:])
    return c
a=list(map(int,input().split()))
b=list(map(int,input().split()))
print(merge(a,b))
```

```python
# Cho ma trận hai chiều nxm có các phần tử được sắp xếp trên
# dòng tăng dần và trên cột tăng dần, hãy tìm số lớn thứ k (k <
# nxm) với thuật toán tốt nhất có thể
k=int(input())
m=int(input())
n=int(input())
matrix=[]
c=[]
for i in range(m):
    hangmoi=[]
    for j in range(n):
        a=int(input())
        hangmoi.append(a)
    matrix.append(hangmoi)
for i in matrix:
    for j in i:
        c.append(j)
c.sort()
c.reverse()
print(c[k-1])
```

```python
# đổi chỗ đúng 2 số để tạo ra số nhỏ nhất có thể
def doicho(a):
    n=list(a)
    b=-1
    for i in range(len(n)):
        for j in range(i+1,len(n)):
            if n[j] < n[i]:
                if i==0:
                    if n[j] !='0' and b==-1:
                        b=j
                    elif b!=-1 and n[j] < n[b]:
                        b=j
                else:
                    if b==-1:
                        b=j
                    else:
                        if n[j] <n[b]:
                            b=j
        if b != -1:
            n[b],n[i]=n[i],n[b]
            break
    return "".join(n)
a=input()
n=list(a)
print(doicho(a))
```
