```python
# Tìm phần tử xuất hiện nhiều hơn n/2 lần(nếu có) với n là số phần tử của 1 dãy số nguyên a
a=map(int,input().split())
b=[]
dic={}
for i in a:
    b.append(i)
n=len(b)/2
for i in b:
    if i not in dic:
        dic[i]=[i]
    else:
        dic[i].append(i)
tim_thay= False
for i in dic:
    e=len(dic[i])
    if e > n/2:
        print("phần tử xuất hiện nhiều hơn n/2 lần là",i)
        tim_thay=True
if not tim_thay:
    print("not found")
```

```python
# Cho 1 số nguyên không âm n do người dùng nhập từ bàn phím. Hãy viết chương trình PY để tính số fibo thứ n theo định nghĩa
def fib(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    return fib(n-1) + fib(n-2)
n=int(input())
print(fib(n))
# vd nhập 3 thì nó sẽ ktra rằng n khác 0 và n khác 1,sau đó trả kết quả cảu fib(2)+fib(1)
# lúc này nó sẽ lại đi tính fib(2)=fib(1)+fib(0)
# mà n=1 thì return 1 và n=0 thì return 0 --> fib(2)=1+0 rồi fib(3)= 1+0+1=2 tương tự với các fibo lớn hơn
```

```python
n = input() # Đọc dưới dạng chuỗi để dễ xử lý từng chữ số
d = [int(i) for i in n] # Chuyển chuỗi thành danh sách các số nguyên

if d:
    # 1. Tìm vị trí của số nhỏ nhất
    vi_tri_min = d.index(min(d))
    
    # 2. Hoán đổi số nhỏ nhất với chữ số đầu tiên (bên trái)
    d[0], d[vi_tri_min] = d[vi_tri_min], d[0]
    
    # 3. In kết quả
    for i in d:
        print(i, end="")
```
